```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV dosyasını oku
veriSeti = pd.read_csv("veriler.csv")

# Ülke seçimi için input al
ulke = "Turkey"

# Seçilen ülkenin verilerini filtrele
ulkeVerisi = veriSeti[veriSeti["Entity"] == ulke]

# Yıl ve abone sayısı verilerini al
yillar = ulkeVerisi["Year"]
aboneSayilari = ulkeVerisi["Fixed broadband subscriptions (per 100 people)"]

# Grafiği oluştur
plt.plot(yillar, aboneSayilari, marker="o")
plt.title(f"Türkiye!!!!!!!!!!! İnternet Kullanımı (Yıllara Göre)")
plt.xlabel("Yıl")
plt.ylabel("Sabit Abonelik Sayıları")
plt.grid(True)
plt.show()
```
