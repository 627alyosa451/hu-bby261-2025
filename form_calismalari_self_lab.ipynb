```python
import ipywidgets as widgets
from IPython.display import display, clear_output
from time import sleep
x = "."
for i in range(0,110,10):
  if i  == 30 or i == 60 or i == 90:
    x = "."
  elif i != 10:
    x += "."
  sleep(0.25)
  clear_output()
  display(widgets.IntProgress(description=f"loading{x}", value=i, min=0, max=100))

clear_output()
display(widgets.IntProgress(description="ready", value=100, min=0, max=100))
```

```python
x = (widgets.ToggleButton(description='click on me!'))
display(x)

def etkin_mi(kontrol):
  print(kontrol)
  if kontrol["new"] == True:
    print("you clicked on me!")

x.observe(etkin_mi, names='value')
```
