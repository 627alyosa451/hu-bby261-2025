```python
import ipywidgets as widgets
from IPython.display import display, clear_output
from time import sleep

slider = widgets.IntSlider(value=1, min=0, max=10)
display(slider)

buton = widgets.Button(description="Click Me!")
display(buton)

progress_bar = widgets.IntProgress(description="doruluk seviyesi", value=0, min=0, max=100)
display(progress_bar)

sifre = [1, 4, 7, "s"]
sifre_index = 0

def buton_kontrol(kontrol):
    global sifre_index

    if slider.value == sifre[sifre_index]:
        print("doru")

        # Geerli ifrenin deeri kontrol ediliyor
        if sifre[sifre_index] == 1:
            progress_bar.value = 33
        elif sifre[sifre_index] == 4:  # 4 iin gncelleme
            progress_bar.value = 66
        elif sifre[sifre_index] == 7:  # 7 iin gncelleme
            progress_bar.value = 100

        sifre_index += 1  # Bir sonraki elemana ge

    if sifre_index < len(sifre) and sifre[sifre_index] == "s":
        print("giri yapld")

def calis(degisken):
    degisken["new"]

buton.on_click(buton_kontrol)
```
