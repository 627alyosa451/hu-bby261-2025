```python
import ipywidgets as widgets
from IPython.display import display, clear_output
from time import sleep

for i in range(0,110,10):
  sleep(0.25)
  clear_output()
  display(widgets.IntProgress(description="loading", value=i, min=0, max=100))
```

```python
x = (widgets.ToggleButton(description='click on me!'))
display(x)

def etkin_mi(kontrol):
  print(kontrol)
  if kontrol["new"] == True:
    print("you clicked on me!")

x.observe(etkin_mi, names='value')
```
