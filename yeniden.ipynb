```python
import requests
from bs4 import BeautifulSoup
```

```python
#URL'NÄ°N "TÄ°TLE" KISMINI BULAN KOD
kullanici = input("Https ekli halde url'yi giriniz: ")
x = requests.get(kullanici).text
#get komutu ile sayfanÄ±n html kodlarÄ±nÄ± x adlÄ± deÄŸiÅŸkene attÄ±m.
soup = BeautifulSoup(x, 'html.parser')
#html kodlarÄ±nÄ± yorumlamasÄ± iÃ§in beautifulsoup a attÄ±m ve bunlarÄ± soup adlÄ± deÄŸiÅŸkene attÄ±m
if soup.title:
  print(soup.title.string)
else:
  print("hata")
#baÅŸlÄ±ÄŸÄ± bulmasÄ±nÄ± saÄŸlayan kodu yazdÄ±m (bu kodlar beautifulsoap adlÄ± sÄ±nÄ±fta tanÄ±mlÄ±)
```

```python
#URL'DEKÄ° DÄ°ÄER LÄ°NKLERÄ° BULAN KOD
for link in soup.find_all("a"): #(linki yukarÄ±daki kod satÄ±rÄ±ndan alÄ±yor) html kodlarÄ±nda "a" geÃ§en yerleri buluyor.("a" geÃ§en yerde bÃ¼yÃ¼k bir metin bulunuyor)
  print(link.get("href")) # bu bÃ¼yÃ¼k metinin iÃ§inden linkleri bu kodla buluyor ("href" kelimesi linklerden hemen Ã¶nce geliyor)
```

```python
#BULUNAN BU LÄ°NKLERE GÄ°RME :,)
for link in soup.find_all("a"): #Ã¼sttekiyle aynÄ± kod karÄ±ÅŸÄ±klÄ±k olmasÄ±n diye tekrar yazdÄ±m
  bulunan = link.get("href")
  if bulunan.startswith("http"):
    y = requests.get(bulunan).text
    soup1 = BeautifulSoup(y, "html.parser")
    if soup1.title:
      print(soup1.title.string) #kabaca Ã§alÄ±ÅŸÄ±yor dÃ¶ngÃ¼ler ve Ã¶ys deki diÄŸer adÄ±mlarÄ± kontrol et
    else:
      print("hata")
```
