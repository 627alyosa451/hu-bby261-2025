```python
import requests
from bs4 import BeautifulSoup
```

```python
#URL'NİN "TİTLE" KISMINI BULAN KOD
kullanici = input("Https ekli halde url'yi giriniz: ")
x = requests.get(kullanici).text
#get komutu ile sayfanın html kodlarını x adlı değişkene attım.
soup = BeautifulSoup(x, 'html.parser')
#html kodlarını yorumlaması için beautifulsoup a attım ve bunları soup adlı değişkene attım
if soup.title:
  print(soup.title.string)
else:
  print("hata")
#başlığı bulmasını sağlayan kodu yazdım (bu kodlar beautifulsoap adlı sınıfta tanımlı)
```

```python
#URL'DEKİ DİĞER LİNKLERİ BULAN KOD
for link in soup.find_all("a"): #(linki yukarıdaki kod satırından alıyor) html kodlarında "a" geçen yerleri buluyor.("a" geçen yerde büyük bir metin bulunuyor)
  print(link.get("href")) # bu büyük metinin içinden linkleri bu kodla buluyor ("href" kelimesi linklerden hemen önce geliyor)
```

```python
#BULUNAN BU LİNKLERE GİRME :,)
for link in soup.find_all("a"): #üsttekiyle aynı kod karışıklık olmasın diye tekrar yazdım
  bulunan = link.get("href")
  if bulunan.startswith("http"):
    y = requests.get(bulunan).text
    soup1 = BeautifulSoup(y, "html.parser")
    if soup1.title:
      print(soup1.title.string) #kabaca çalışıyor döngüler ve öys deki diğer adımları kontrol et
    else:
      print("hata")
```
