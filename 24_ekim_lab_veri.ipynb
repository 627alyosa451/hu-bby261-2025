```python
import pandas as pd
from datetime import datetime
import csv
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

headers = ['Sensor Value','Date','Time']
# Tell pandas that the first row contains headers
df = pd.read_csv('/content/broadband-penetration-by-country.csv', names=headers, header=0)
print (df)

# Now, the 'Date' column should only contain date strings, not the header
df['Date'] = df['Date'].map(lambda x: datetime.strptime(str(x),"%Y")) # Changed format to match your data
x = df['Date']
y = df['Sensor Value']

# plot
plt.plot(x,y)
# beautify the x-labels
plt.gcf().autofmt_xdate()

plt.show()
```
